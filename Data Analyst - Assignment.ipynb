```python
#function to covert weights as per requirement

def round(x):
    if x < 0.500 :
        return '0.5'
    elif x<1 and x>0.5:
        return '1'
    elif x<1.500 and x>1:
        return '1.5'
    elif x<2 and x>1.500:
        return '2'
    elif x<2.500 and x>2:
        return '2.5'
    elif x<3 and x>2.500:
        return '3'
    elif x<3.500 and x>3:
        return '3.5'
    elif x<4 and x>3.500:
        return '4'
    elif x<4.5 and x>4:
        return '4.5'
    else:
        return x
```

```python
#applying functiont to our feature for next process      
final['Weight slab as per X (KG)']=final['Total weight as per X (KG)'].apply(lambda x:round(x))
final['Weight slab as per X (KG)'].unique()
```
